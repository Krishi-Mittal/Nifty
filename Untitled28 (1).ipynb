```python
nifty['date'] = pd.to_datetime(nifty['date'], format='%Y-%m-%d %H:%M:%S', errors='coerce') #Converted date variable to datetime for analysis
nifty.set_index('date', inplace=True)
```

```python
#Plotting Closing price against date from 2015 to 2024
plt.figure(figsize=(12,6))
plt.plot(nifty.index, nifty['close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('NIFTY 100 Close Price')
# Converted year values to integers using astype(int) to avoid the TypeError
for year in range(int(min(nifty.index.year)), int(max(nifty.index.year))+1):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--')
plt.show()
```

The following graph clearly shows that the time series data is not stationary. So we will take first difference and then test stationarity using Augmented Dickey Fuller Test(ADF Test).

```python
daily_data = nifty.resample('D').mean() #Have indexed minutes data to single value for each day using mean
daily_data['close_diff'] = daily_data['close'].diff().dropna() #Have dropped null vaalues, if any
adf_test = adfuller(daily_data['close_diff'].dropna()) #Performed ADF test to check stationarity
print("ADF Statistic:", adf_test[0])
print("p-value:", adf_test[1])
print("Critical Values:", adf_test[4])
```

The ADF Test statistic value is less than critical values at all level of significance. Also, P value is less than 0.01, 0.05 and 0.1. So we rejct null hypothesis and thus, the time series data is stationary now.

```python
#Plotted Nifty closing price from 2015 to 2024 after making it stationary by taking first difference
plt.figure(figsize=(12,6))
plt.plot(daily_data.index, daily_data['close_diff'])
plt.xlabel('Date')
plt.ylabel('Daily Close Price Difference')
plt.title('NIFTY 100 Daily Close Price Difference')
for year in range(min(daily_data.index.year), max(daily_data.index.year)+1):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--')
```

```python
plt.figure(figsize=(12, 6))

# ACF Plot
plt.subplot(2, 1, 1)
plot_acf(daily_data['close_diff'].dropna(), lags=40, ax=plt.gca())
plt.title("ACF (Autocorrelation Function)")

# PACF Plot
plt.subplot(2, 1, 2)
plot_pacf(daily_data['close_diff'].dropna(), lags=40, ax=plt.gca(), method="ywm")
plt.title("PACF (Partial Autocorrelation Function)")

plt.tight_layout()
plt.show()
```

The following plot shows only 1 significant spike for both ACF and PACF. Thus the order for AR model (p lags is 1 using PACF) and for MA model (q lags is 1 using ACF).

```python
# Defined the size of the training set (80% of the data)
train_size = int(len(daily_data) * 0.8)

# Splitted the data into train and test sets
train = daily_data['close'][:train_size].dropna()
test = daily_data['close'][train_size:].dropna()
```

```python

model = ARIMA(train, order=(1, 1, 1))  # Adjusted order=(p, d, q) based on ACF/PACF plots
model_fit_1 = model.fit()

# Summary of the model
print(model_fit_1.summary())

model = ARIMA(train, order=(1, 1, 2))  # Adjusted order=(p, d, q) based on ACF/PACF plots
model_fit_2 = model.fit()

# Summary of the model
print(model_fit_2.summary())

model = ARIMA(train, order=(2, 1, 1))  # Adjusted order=(p, d, q) based on ACF/PACF plots
model_fit_3 = model.fit()

# Summary of the model
print(model_fit_3.summary())
```

All the co-efficients (AR1,MA1, Sigma2) have p value less than 0.05 thus all are significant. And AIC< BIC, HQIC results shows ARMA(1,1,1) as the best model among all.
Q Statistic has p value < 0.05, thus residuals may have autocorrelation.
H test suggests Variance is not constant.
JB test suggests residuals are normally distributed.

The p value of less than o.o5 indicates that we reject null hypothesis, thus there is arch affect and therefore we should try GARCH Modelling.


```python
import statsmodels.api as sm

sm.qqplot(model_fit_1.resid, line="s")
plt.show()

# Changed plt.acf to plot_acf from statsmodels.graphics.tsaplots
from statsmodels.graphics.tsaplots import plot_acf  # Import plot_acf
plot_acf(model_fit_1.resid)  # Use plot_acf instead of plt.acf
plt.show()
```

The red line shows normal distribtuion, whereas, blue dosts are our residuals and they dont align so they are not normally distributed, but for better assurance, we will perform JB test.

```python
from statsmodels.stats.stattools import jarque_bera

jb_test = jarque_bera(model_fit_1.resid)
print("Jarque-Bera Test Statistic:", jb_test[0])
print("p-value:", jb_test[1])
```

Since p value is < o.o5, the residuals are not nomrally distribued, thus we will try GARCH or EGARCH

Since p value is less tha 0.05, there is significant hetroscedasticity, thus it again confirms to move to GARCH.

```python
from statsmodels.stats.diagnostic import het_arch

residuals = model_fit_1.resid
h_test = het_arch(residuals)
print("H Test Statistic:", h_test[0])
print("p-value:", h_test[1])
```

Since p value < 0.05, there is presence of hetroscedasticity and thus we move to GARCH Modelling.

```python
garch_model = arch_model(residuals, vol='GARCH', p=1, q=1)
garch_fit = garch_model.fit(disp="off")
print(garch_fit.summary())
```

```python
egarch_model = arch_model(residuals, vol='EGARCH', p=1, q=1)
egarch_fit = egarch_model.fit(disp="off")
print(egarch_fit.summary())
```

Since GARCH Modelling has lower AIC and BIC along with higher Log Likelihood than EGARCH, we move forward with GARCH Modelling.

```python
forecast = garch_fit.forecast(horizon=len(test))
forecasted_volatility = forecast.variance.iloc[-1]
print("Forecasted Volatility:", forecasted_volatility)
```

```python
split = int(len(residuals) * 0.8) #Wpliited residuals into train and test with 80% train and 20% test data
train = residuals[:split]
test = residuals[split:]
```

```python
garch_model = arch_model(train, vol='GARCH', p=1, q=1) #Trained GARCH Model on trained data
garch_fit = garch_model.fit(disp="off")
print(garch_fit.summary())
```

```python
forecast = garch_fit.forecast(horizon= 30)
print(forecast.mean[-1:])
print(forecast.variance[-1:])
```

```python
history = list(train)
predictions = []

for t in range(len(test)):
    model = arch_model(history, vol='GARCH', p=1, q=1)
    model_fit = model.fit(disp="off")
    forecast = model_fit.forecast(horizon=1)
    # Accessing the forecasted variance for the last time step
    pred_var = forecast.variance.iloc[-1, 0]
    predictions.append(np.sqrt(pred_var))

    history.append(test.iloc[t])
```

```python
#Plotted Actuald and Futue Volatility
plt.figure(figsize=(12, 6))
plt.plot(test.index, np.abs(test), label='Actual Volatility', color='blue')
plt.plot(test.index, predictions, label='Predicted Volatility', color='red')
plt.title('Rolling Forecast - GARCH Model')
plt.legend()
plt.show()
```

As evident from the plot, the model fits well as the spikes correspond with each other. Under Garch Modelling the Magnitude shouldnot be same, but the pattern should be similar.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(np.abs(test), predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(np.abs(test), predictions)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
```

The RMSE and MAE, also tells the model fits well with the data.

```python
future_forecast = garch_fit.forecast(horizon=30)
future_forecasted_volatility = pd.Series(np.sqrt(future_forecast.variance.iloc[-1]))
print("Forecasted Volatility for the next 30 days:", future_forecasted_volatility)
```

```python
#Plotted Forecasted Volatility for next 30 days
plt.figure(figsize=(12, 6))
plt.plot(pd.date_range(start=test.index[-1], periods=30, freq='D'), future_forecasted_volatility, label='Forecasted Volatility', color='orange')
plt.title('Forecasted Volatility for the Next 30 Days')
plt.xlabel('Date')
plt.ylabel('Forecasted Volatility')
plt.show()
```
